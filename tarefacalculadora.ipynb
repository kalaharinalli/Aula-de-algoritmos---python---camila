```python
import math

# Funções para as operações básicas
def adicionar(x, y):
    return x + y

def subtrair(x, y):
    return x - y

def multiplicar(x, y):
    return x * y

def dividir(x, y):
    return x / y

def potencia(x, y):
    return x ** y

def raiz_quadrada(x):
    return math.sqrt(x)

# Dicionário que mapeia as escolhas para as funções
operacoes = {
    '1': adicionar,
    '2': subtrair,
    '3': multiplicar,
    '4': dividir,
    '5': potencia,
    '6': raiz_quadrada
}

# Mostrar o menu de opções
print("Selecione a operação:")
print("1 - Adição")
print("2 - Subtração")
print("3 - Multiplicação")
print("4 - Divisão")
print("5 - Potência")
print("6 - Raiz Quadrada")

# Pedir ao usuário para escolher uma operação
escolha = input("Digite sua escolha (1/2/3/4/5/6): ")

# Pedir ao usuário para digitar os números conforme a operação escolhida
if escolha in ['1', '2', '3', '4', '5']:
    num1 = float(input("Digite o primeiro número: "))
    num2 = float(input("Digite o segundo número: "))
    resultado = operacoes[escolha](num1, num2)
    print(f"O resultado é: {resultado}")
elif escolha == '6':
    num = float(input("Digite o número: "))
    resultado = operacoes[escolha](num)
    print(f"O resultado é: {resultado}")
else:
    print("Escolha inválida")
```

```python
# Definir as funções para as operações básicas
def adicionar(x, y):
    return x + y

def subtrair(x, y):
    return x - y

def multiplicar(x, y):
    return x * y

def dividir(x, y):
    return x / y

# Dicionário que mapeia as escolhas para as funções
operacoes = {
    '1': adicionar,
    '2': subtrair,
    '3': multiplicar,
    '4': dividir
}

# Mostrar o menu de opções
print("Selecione a operação.")
print("1 - Adição")
print("2 - Subtração")
print("3 - Multiplicação")
print("4 - Divisão")

# Pedir ao usuário para escolher uma operação
escolha = input("Digite sua escolha (1/2/3/4): ")

# Pedir ao usuário para digitar dois números
num1 = float(input("Digite o primeiro número: "))
num2 = float(input("Digite o segundo número: "))

# Verificar se a escolha é válida e realizar a operação
if escolha in operacoes:
    # Verificar se a operação é divisão e o divisor não é zero
    if escolha == '4' and num2 == 0:
        print("Erro! Divisão por zero não é permitida.")
    else:
        # Chamar a função correspondente à escolha do usuário
        resultado = operacoes[escolha](num1, num2)
        print(f"O resultado da operação é: {resultado}")
else:
    print("Escolha inválida")
```
